```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# Carrega a imagem colorida
imagem_colorida = cv2.imread("elcid.jpg")

# Converte de BGR (OpenCV) para RGB (matplotlib)
imagem_rgb = cv2.cvtColor(imagem_colorida, cv2.COLOR_BGR2RGB)

# Converte para tons de cinza
imagem_cinza = cv2.cvtColor(imagem_colorida, cv2.COLOR_BGR2GRAY)

# Binariza com limiar
limiar = 127
_, imagem_binarizada = cv2.threshold(imagem_cinza, limiar, 255, cv2.THRESH_BINARY)
```

```python
# Exibe as 3 verses lado a lado
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(imagem_rgb)
plt.title('Imagem Colorida')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(imagem_cinza, cmap='gray')
plt.title('Imagem em Tons de Cinza')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(imagem_binarizada, cmap='gray')
plt.title('Imagem Binarizada (P&B)')
plt.axis('off')

plt.show()
```
